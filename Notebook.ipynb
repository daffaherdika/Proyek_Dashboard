### Cleaning Data

Membersihkan data product

```python
product_df[product_df.product_category_name.isna()]
```

```python
product_df = product_df.dropna(subset=['product_category_name',
                           'product_name_lenght',
                           'product_description_lenght',
                           'product_photos_qty'])
```

```python
product_df = product_df[product_df['product_weight_g'] > 0]
```

```python
product_df.info()
product_df.isna().sum()
```

Membersihkan data pada order_items_df

```python
datetime_columns = ["shipping_limit_date"]

for column in datetime_columns:
  order_items_df[column] = pd.to_datetime(order_items_df[column])

order_items_df.info()
```

Membersihkan data pada orders_df

```python
orders_df = orders_df.dropna()
```

```python
orders_df['order_purchase_timestamp'] = pd.to_datetime(orders_df['order_purchase_timestamp'])
orders_df['order_delivered_customer_date'] = pd.to_datetime(orders_df['order_delivered_customer_date'])
orders_df['order_delivered_carrier_date'] = pd.to_datetime(orders_df['order_delivered_carrier_date'])
orders_df['order_estimated_delivery_date'] = pd.to_datetime(orders_df['order_estimated_delivery_date'])
```

```python
orders_df.info()
orders_df.isna().sum()
```

Membersihkan data pada payment

```python
payment_df[payment_df['payment_value'] > 0]
```

```python
payment_df = payment_df[payment_df['payment_type'] != 'not_defined']
```

## Exploratory Data Analysis (EDA)

### Explore ...

Untuk mengetahui kota mana dengan jumlah pesanan terbanyak

```python
df_customer_orders = orders_df.merge(customers_df, on="customer_id", how="inner")
top_cities = df_customer_orders['customer_city'].value_counts().head(10).reset_index()
top_cities.columns = ['Kota', 'Jumlah Pesanan']
top_cities.head(10)
```

Untuk mengetahui Produk kategori yang paling banyak terjual

```python
df_orders_products = order_items_df.merge(product_df, on="product_id", how="left")
top_categories = df_orders_products['product_category_name'].value_counts().reset_index()
top_categories.columns = ['Kategori Produk', 'Jumlah Terjual']
top_categories.head(10)
```

Untuk mengetahui rata-rata waktu pengiriman pesanan ke pelanggan

```python
orders_df['delivery_time_days'] = (orders_df['order_delivered_customer_date'] - orders_df['order_purchase_timestamp']).dt.days
avg_delivery_time = orders_df['delivery_time_days'].mean()
avg_delivery_time
```

Untuk mengetahui rata-rata nilai transaksi berdasarkan metode pembayaran

```python
avg_payment_per_type = payment_df.groupby('payment_type')['payment_value'].mean().reset_index()
avg_payment_per_type.columns = ['Metode Pembayaran', 'Rata-rata Nilai Transaksi']
avg_payment_per_type_sorted = avg_payment_per_type.sort_values(by="Rata-rata Nilai Transaksi", ascending=False)
avg_payment_per_type_sorted
```

```python
payment_counts = payment_df['payment_type'].value_counts().reset_index()
payment_counts.columns = ['Metode Pembayaran', 'Jumlah Transaksi']
payment_counts
```

**Insight:**
- **Pertanyaan 1**:  kota mana dengan jumlah pesanan terbanyak?
  **Jawaban:** kota sao paulo dengan 15540 pesanan
- **Pertanyaan 2**: Produk kategori apa yang paling banyak terjual?
  **Jawaban:** Kategori cama_mesa_banho adalah produk paling laris dengan 11.115 penjualan.
- **Pertanyaan 3:** Berapa rata-rata waktu yang dibutuhkan untuk mengirim pesanan ke pelanggan?
  **Jawaban:** Rata-rata pesanan sampai dalam 12 hari
- **Pertanyaan 4:** Berapa rata-rata nilai transaksi dari setiap metode pembayaran?
  **Jawaban:** Kartu kredit, dengan nilai transaksi rata-rata tertinggi sebesar 163,32
- **Pertanyaan 5**:  Metode pembayaran apa yang paling sering digunakan?
  **Jawaban:** Metode pembayaran credit_card dengan 76795 transaksi

## Visualization & Explanatory Analysis

### Pertanyaan 1: kota mana dengan jumlah pesanan terbanyak?

```python
plt.figure(figsize=(10, 5))
sns.barplot(
    x=top_cities['Kota'],
    y=top_cities['Jumlah Pesanan'],
    hue=top_cities['Kota'],
    palette='Blues_d',
    legend=False
)

plt.xticks(rotation=45)
plt.xlabel('Kota')
plt.ylabel('Jumlah Pesanan')
plt.title('Kota dengan Jumlah Pesanan Terbanyak')
plt.show()
```

### Pertanyaan 2: Produk kategori apa yang paling banyak terjual?

```python
plt.figure(figsize=(10, 5))
sns.barplot(
    x=top_categories['Kategori Produk'][:10],
    y=top_categories['Jumlah Terjual'][:10],
    hue=top_categories['Kategori Produk'][:10],
    palette='Greens_d',
    legend=False
)

plt.xticks(rotation=45)
plt.xlabel('Kategori Produk')
plt.ylabel('Jumlah Terjual')
plt.title('Kategori Produk Paling Banyak Terjual')
plt.show()
```

### Pertanyaan 3: Berapa rata-rata waktu yang dibutuhkan untuk mengirim pesanan ke pelanggan?

```python
plt.figure(figsize=(8, 5))
sns.histplot(orders_df['delivery_time_days'].dropna(), bins=20, kde=True, color='purple')
plt.xlabel('Waktu Pengiriman (Hari)')
plt.ylabel('Frekuensi')
plt.title('Distribusi Waktu Pengiriman Pesanan')
plt.show()
```

### Pertanyaan 4: Berapa rata-rata nilai transaksi dari setiap metode pembayaran?

```python
plt.figure(figsize=(10, 5))
sns.barplot(data=avg_payment_per_type_sorted,
            x="Metode Pembayaran",
            y="Rata-rata Nilai Transaksi",
            hue="Metode Pembayaran",
            palette='OrRd',
            legend=False)

plt.xlabel('Metode Pembayaran')
plt.ylabel('Rata-rata Nilai Transaksi (BRL)')
plt.title('Rata-rata Nilai Transaksi Berdasarkan Metode Pembayaran')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
sns.barplot(
    x=payment_counts['Metode Pembayaran'],
    y=payment_counts['Jumlah Transaksi'],
    hue=payment_counts['Metode Pembayaran'],  # Tambahkan hue
    palette='coolwarm',
    legend=False  # Supaya legend tidak muncul
)
plt.xlabel('Metode Pembayaran')
plt.ylabel('Jumlah Transaksi')
plt.title('Metode Pembayaran Paling Sering Digunakan')
plt.show()
```

### Pertanyaan 5: Metode pembayaran apa yang paling sering digunakan?

```python
fig, ax = plt.subplots(figsize=(10, 5))
sns.barplot(
    x=payment_counts['Metode Pembayaran'],
    y=payment_counts['Jumlah Transaksi'],
    hue=payment_counts['Metode Pembayaran'],  # Tambahkan hue
    palette='coolwarm',
    legend=False  # Supaya legend tidak muncul
)
plt.xlabel('Metode Pembayaran')
plt.ylabel('Jumlah Transaksi')
plt.title('Metode Pembayaran Paling Sering Digunakan')
plt.show()
```

**Insight:**
- Sao Paulo menjadi kota dengan jumlah pesanan paling banyak
- Kartu Kredit adalah metode pembayaran dengan rata-rata transaksi tertinggi dan yang paling sering digunakan
- Metode boleto & debit_card memiliki nilai transaksi yang lebih rendah dibanding kartu kredit
- Metode voucher memiliki rata-rata transaksi yang paling rendah
